```python
class One_Stop_Coffee():
    
   #This will be the beginning of the chatbot taking the order of our customer at the coffee shop.
   #They will begin by welcoming in the customer, and then ask the customer for the name to be called for their order. 
    def __init__(self):
        pass
                 
    def welcome(self):
        #This is the welcoming statement, in which the customer is welcomed to the coffee shop.
        #The customer is then asked for the name which belongs to their order.
        print("\033[1;35;47m\n")
        print("\nWelcome to Linnea's One Stop Coffee Shop!")
        name = input('\nMay I have a name for your order?')
        print("\nThanks for coming in today, " + name + ".")
        customer_feeling = input("\nHow are you doing today?")
        print("Thanks for sharing!")
    
    def take_order(self):
        """Take the order of the customer accurately.
         Parameters 
        ----------
        None
        
         Returns
        ----------
        None
        
        Print Statement : str
            This statement prints out the drink elements into a statement of the order. 
        """
        drink, milk, flavor = self.drink_type()
        size = self.drink_size()
        temperature = self.drink_temp()
        straw = self.straw_need()
        pastry = self.pastry_order()
        prices = self.determine_price()
        #This is where all of the subsequent functions are called, in order to take orders.
        
        if drink == 'Latte':
            print("Sounds great, friend! Your drink order today is a"
                  " " + size + ' ' + temperature + ' ' + flavor + ' ' + milk + ' ' + drink + ' ' + straw + 
                  ". Your food order is a " + pastry +  '. Your total will be $'
                  + str(self.prices) + '. You may wait in the cafe until your order is ready.')
        elif drink == 'Americano':
            print("Sounds great, friend! Your drink order today is a"
                  " " + size + ' ' + temperature + ' ' + flavor + ' ' + drink + ' ' + milk + ' and ' + straw + 
                  ". Your food order is a " + pastry + '. Your total will be $' + str(self.prices) + 
                  '. You may wait in the cafe until your order is ready.')
        else:
            print("Sounds great, friend! Your drink order today is a"
                  " " + size + ' ' + temperature + ' ' + drink + ' ' + straw + 
                  ". Your food order is a " + pastry + '. Your total will be $' + str(self.prices) + 
                  '. You may wait in the cafe until your order is ready.')
                
               
    def drink_size(self):
        """ Determine the size preference of the customer for their drink. 
       
       Returns 
       ----------
       'small'/'medium'/ 'large' : str
           Specifies that the customer would like a small/medium/large drink.
        """
       
        size_res = input("\nWhat size would you like for your"
                         " drink today?\n[a] Small \n[b] Medium \n[c] Large")
        
        if size_res == 'a':
            self.size = 'small'
        elif size_res == 'b':
            self.size = 'medium'
        elif size_res == 'c':
            self.size = 'large'
        return self.size

    
    def drink_type(self):
        """This function allows the user to select which drink type they desire.
        Parameters
        ----------
        None
        
        Returns 
        ----------
        self.drink : str
            This specifies which type of drink the customer would like
        self.milk : str
            This specifies what kind of milk the customer will have with their drink. (Standard for all drinks besides
            americanos and lattes is going to be 2% milk)
        self.flavor : str
            This specifies any type of flavor the customer would like. (Standard for drinks that already have flavor
            is regular)
        """
        response = input("\nWhat type of drink would you like today?"
                         "\n[a] Caramel Macchiato \n[b] Latte \n[c] Americano \n[d] Mocha"
                         "\n[e] Cortado \n[f] Espresso")
        if response == 'a':
            self.drink = 'Caramel Macchiato'
            self.milk = '2% milk'
            self.flavor = 'regular'
        elif response == 'b':
            self.drink = 'Latte'
            #In the case of lattes, milk and flavor will be determined in the subsequent latte function.
            self.milk, self.flavor =  self.latte_order()
        elif response == 'c':
            self.drink = 'Americano'
            #In the case of americanos, milk and flavor will be determined in the subsequent americano function.
            self.milk, self.flavor = self.americano_order()
        elif response == 'd':
            self.drink = 'Mocha'
            self.milk = '2% milk'
            self.flavor = 'regular'
        elif response == 'e':
            self.drink = 'Cortado'
            self.milk = '2% milk'
            self.flavor = 'regular'
        elif response == 'f':
            self.drink = 'Espresso'
            self.milk = '2% milk'
            self.flavor = 'regular'
        return self.drink, self.milk, self.flavor
    
    def latte_order(self):
        """This function is called if a customer orders a latte. It specifies what type of milk/flavor the customer
        would like. 
        
        Parameters 
        ----------
        None
        
        Returns 
        ----------
        self.milk : str
            This is the type of alternative milk that the customer selects.
        self.flavor : str
            In this function, it is determined whether the customer would like flavor or not. 
        
        """
        
        milk_res = input("\nWhat kind of milk would you like in your latte?"
                         "\n[a] Oatmilk \n[b] Almondmilk \n[c] 2% milk \n[d] Soymilk"
                         "\n[e] Coconutmilk \n[f] Whole Milk")
                 
        if milk_res == 'a':
            self.milk = 'Oatmilk'
        elif milk_res == 'b':
            self.milk = 'Almondmilk'
        elif milk_res == 'c':
            self.milk = '2% milk'
        elif milk_res == 'd':
            self.milk = 'Soymilk' 
        elif milk_res == 'e':
            self.milk = 'Coconutmilk'
        elif milk_res == 'f':
            self.milk = 'Whole Milk'   
        flavor_res = input("\nWould you like to add flavor to your latte? \n[a] Yes \n[b] No")
        #This is asking if the customer would like to add any flavor.
        
        if flavor_res == 'a':
            self.flavor = self.flavor_selection()
            #If the customer decides they would like to add a flavor, they are taken to the flavor selection function.
        elif flavor_res == 'b':
            self.flavor = 'regular'
            #If the customer decides they do not want to add a flavor, the flavor is stored as 'regular'.
        return self.milk, self.flavor
    
    def flavor_selection(self):
        """This function works to find what kind of flavor the customer would like in their latte.
        Parameters
        ----------
        None
        
        Returns
        ----------
        self.flavor : str
            This returns the specific flavor the customer would like in their latte.
        
        """
        flavor_res_two = input("\nWhat type of flavor would you like in your latte?"
                               "\n[a] Hazelnut \n[b] Toffeenut \n[c] Vanilla \n[d] Cinnamon Dolce"
                               "\n[e] Irish Cream \n[f] Caramel \n[g] White Mocha \n[h] Raspberry")
        if flavor_res_two == 'a':
            self.flavor = 'Hazelnut'
        elif flavor_res_two == 'b':
            self.flavor = 'Toffeenut'
        elif flavor_res_two == 'c':
            self.flavor = 'Vanilla'
        elif flavor_res_two == 'd':
            self.flavor = 'Cinnamon Dolce'
        elif flavor_res_two == 'e':
            self.flavor = 'Irish Cream'
        elif flavor_res_two == 'f':
            self.flavor = 'Caramel'
        elif flavor_res_two == 'g':
            self.flavor = 'White Mocha'
        elif flavor_res_two == 'h':
            self.flavor = 'Raspberry'
        return self.flavor
        
        
    def americano_order(self):
        """This function determines the milk and flavor options for an americano.
        
        Parameters 
        ----------
        None
        
        Returns
        ----------
        self.milk : str
            This indicates whether the customer would like cream in their americano or not.
        self.flavor : str
            This stores 'regular', which is the baseline for all of the drinks (other than lattes).
        """
        cream_res = input("\nWould you like any creamer in your Americano? \n[a] Yes \n[b] No")
        #the flavor for americanos is always regular, like the other drinks.
        if cream_res == 'a':
            self.milk = 'with cream'
            self.flavor = 'regular'
        elif cream_res == 'b':
            self.cream = 'without cream'
            self.flavor = 'regular'
        return self.milk, self.flavor
        
    def drink_temp(self):
        """This function works to determine the temperature of the customer's drink. 
        Parameters
        ----------
        None
        
        Returns
        ----------
        'hot' : str
            This specifies that the customer wants a hot drink.
        'iced' : str 
            This specifies that the customer wants a cold drink.
        """
        d_temp = input("\nWould you like your drink hot or iced today?\n[a] Hot \n[b] Iced")
        if d_temp == 'a':
            return 'hot'
        elif d_temp == 'b':
            return 'iced'
          
    
    def straw_need(self):
        """This function determines whether the customer would like a straw with their order.
        
        Parameters 
        ----------
        None
        
        Returns 
        ----------
        'with a straw' : str
            This indicates that the customer would like a straw.
        'without a straw' : str
            This indicates that the customer would not like a straw. 
        """
        straw_res = input("\nWould you like a paper straw with your drink today? \n[a] Yes \n[b] No")
        if straw_res == 'a':
            return 'with a straw'
        elif straw_res == 'b':
            return 'without a straw'
        
    
    def pastry_order(self):
        """This function works to determine which pastry a customer would like for free. 
        
        Parameters 
        ----------
        None
        
        Returns
        ----------
        self.pastry : str
            This indicates which pastry the customer would like. 
        """
        pastry_res_two = input("\nWe are having an Opening Day special today! All pastries are on the house."
                               " What kind of pastry would you like to pair with your drink today? \n[a] Chocolate Croissant"
                           "\n[b] Almond Croissant \n[c] Cheese Danish \n[d] Double Fudge Brownie \n[e] Ham and Swiss Croissant")
        if pastry_res_two == 'a':
            self.pastry = 'Chocolate Croissant'
        elif pastry_res_two == 'b':
            self.pastry = 'Almond Croissant'
        elif pastry_res_two == 'c':
            self.pastry = 'Cheese Danish'
        elif pastry_res_two == 'd':
            self.pastry = 'Double Fudge Brownie'
        elif pastry_res_two == 'e':
            self.pastry = 'Ham and Swiss Croissant'
        return self.pastry
        
    def determine_price(self):
        """This function works to determine the price of the drinks. 
        
        Parameters
        ----------
        None
        
        Returns 
        ---------
        self.prices : int
            This is the price of the drink when all factors (milk, flavor, drink type) are considered.
        """
        drink_price = {'Caramel Macchiato' : 4.50, 'Latte' : 4.00, 'Americano' : 3.50, 'Mocha': 4.50, 'Cortado' : 3.00,
                       'Espresso' : 2.50}
        milk_price = {'Oatmilk' : .50, 'Almondmilk' : .50, '2% milk' : 0, 'Soymilk' : .50, 'Coconutmilk' : .50,
                      'Whole Milk' : 0, 'with cream' : 0, 'without cream' : 0}
        flavor_price = {'Hazelnut' : .50, 'Toffeenut' : .50, 'Vanilla' : .50, 'Cinnamon Dolce' : .50,
                        'Irish Cream' : .50, 'Caramel' : .50, 'White Mocha' : .50, 'Raspberry' : .50, 'regular' : 0}
        self.prices = milk_price[self.milk] + drink_price[self.drink] + flavor_price[self.flavor]
        #This is where the prices of each category (drink, milk, and flavor) are all added. 
        return self.prices
   
   
```
